```python
#multi linear regression model
```

```python
x = new_df[['TOEFL Score','CGPA','University Rating','GRE Score','Research']] #independent 
y = new_df['Chance of Admit '] #dependent variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# create a Linear Regression
lm = LinearRegression()
```

```python
# fit the model
lm.fit(X_train,y_train)
```

```python
# predict the model for test 
predictions = lm.predict(X_test)
```

```python
# Feature scaling (Bringing whole data into a single form)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(X_train)
x_test = sc.transform(X_test)
```

```python
print('Comparing actual and predicted values')
print('Predicted values are almost similar to actual values')
np.column_stack((predictions, y_test))
```

```python
# plot the model to see if it fits the testing data
plt.figure(figsize = (4, 3))
plt.scatter(y_test, predictions)
plt.xlabel('Y_test')
plt.ylabel('Prediction of LinearRegression')
plt.title("Actual Data vs Model Prediction")
plt.show()
```

```python
from sklearn import metrics
```

```python
print('Percentage for Train Data: {}'.format(lm.score(X_train, y_train)))
print('Percentage for Test Data: {}'.format(lm.score(X_test, y_test)))
```

```python
print('Model is accurate because value of regression metrics is close to 0. Hence, the model created is performing well.')
print('\n')
print('1. MAE:', metrics.mean_absolute_error(predictions, y_test))
print('MAE is absolute difference between actual and predicted values and sum all the errors and divide them by a total number of observation')
print('MAE is very less that means model is accurate and have less errors')
print('\n')
print('2. MSE:', metrics.mean_squared_error(predictions, y_test))
print('MSE is squared distance between actual and predicted values and divide them by a total number of observation')
print('\n')
print('3. RMSE:', np.sqrt(metrics.mean_squared_error(predictions, y_test)))
print('RMSE is square root of mean squared error')
print('RMSE value is as low as 0.06 and therefore the model is better with its predictions')
```

```python
print('1. MAE:', metrics.mean_absolute_error(predictions, y_test))
print('2. MSE:', metrics.mean_squared_error(predictions, y_test))
print('3. RMSE:', np.sqrt(metrics.mean_squared_error(predictions, y_test)))
```
